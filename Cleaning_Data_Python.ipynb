## Part II: Data-Cleaning - The Plan

### Propose:

The plan for cleaning data will begin with checking the data for missing values and replacing data
where possible by imputing or assigning a variable to the data where possible. For instance, if the
value is of a numeric type, it can be replaced with the mean or average of the data around it, so it
does not influence the results. Otherwise, the row can be dropped if the record it too incomplete to
be usable or it is categorical data that would be relevant to the results when using it.

### Approach:

Qualifying the data into either categorical or continuous is an important first step to find any
weaknesses in the dataset. Simple statistical confidence intervals can then be used to determine the
frequency of missing values for either datatype. Any column that is not a numerical value type will
be identified as type “object” to delineate the fact it is categorical vs some other numerical type. In
the first stage of this observation the missing values can be returned as a Boolean value of “True” by
checking if the variable is NULL via a Python method. Then using a predication of this method by
dividing it by the Data Frame length will yield a confidence interval for columns containing null data
as a percentage of missing data per column.

### Programming Language and Libraries:

Python will be used for the project because it can be applied inside a Jupiter Notebook and can
easily verify the results of changes made to the dataset. Pandas and NumPy will be used to import,
display, and manipulate data using a Data Frame of the CSV file. Since cleaning is the primary
objective, the compatibility of Python across platforms will be more suitable for future updates to
the project, as R is more system dependent than Python, and the results will be easy to display via
PDF file or as a HTML page on a company intranet, for example

### Methodology:

Since categorical data is not quantitative as is, an average of the variables cannot be substituted in
place of missing variables. Dropping the missing categorical fields results in 70% of data remaining
for study in the dataset without the errors occurring from missing values of category data. (Bex T,
2020) However this will also drop continuous data as well. Another approach would be to fill in the
missing categorical variables with the string “Missing Data” as an identifier. This is accomplished
with: <code>df[category_data] = df[category_data].fillna('Missing Data')</code> 


```python
# replace missing category data
df[category_data] = df[category_data].fillna('Missing Data')
```

To drop the missing variables (all the Null values), using Pandas to locate them, in the categorical
data could have been accomplished with the following code, but will not be used in this case study:
<code>df[category_data]. = df[category_data].dropna(how='any',axis=0)</code> (Moderc, 2017)

Since the numerical data in the remaining dataset can be imputed or replaced with the mean of the
data in general, this process will be used to correct the remaining missing numeric data as it uses
the central tendency of data. This task is accomplished with:
<code>df[numeric_data] = df[numeric_data].fillna(df.mean())</code> (Kumar, 2020)

```python
# replace null values with mean
df[numeric_data] = df[numeric_data].fillna(df.mean())
```

## Principle Component Analysis:
    
### Principal Components of the Data

To begin the principal component analysis, the 23 variables of the <code>df[numeric_data]</code> data frame
were extracted from the df dataset and normalized. All the components were extracted at this
point because it is not understood how many are going to be useful components. The resulting
scree plot identifies how variable reduction needs to be performed to gain precision in the data
using the sklearn.decomposition package from Python.


```python
# number of rows and columns in the dataset
df[numeric_data].shape
```

```python
# numeric data normalized
from sklearn.decomposition import PCA
df_normalized=(df[numeric_data]-df[numeric_data].mean())/df[numeric_data].std()
pca = PCA(n_components=df[numeric_data].shape[1])
```

```python
# scree plot displaying the normalized dataset

pca.fit(df_normalized)
df_pca = pd.DataFrame(pca.transform(df_normalized), 
                      columns=['PC1','PC2','PC3','PC4','PC5','PC6','PC7','PC8',
                                                             'PC9','P10','P11','P12','P13','P14','P15','P16','P17',
                                                             'P18','P19','P20','P21','P122','P23'])
                     
plt.plot(pca.explained_variance_ratio_)
plt.xlabel('Number of Components')
plt.ylabel('Explained Variance')
plt.show()
```

### Identification of Principal Components:
    
Next the extracted principal components were converted to eigenvalues to explain variance in the
set. Eigenvalues in the graph below that are greater than 1 are better than the average. This means
the first 10 principle components contain the most presision among the rest within the data.

```python
from numpy.linalg import eig

# convert dataset to eigenvalues
cov_matrix = np.dot(df_normalized.T, df_normalized) / df.shape[0]

eigenvalues,eigenvectors=eig(cov_matrix)
plt.plot(eigenvalues)
plt.xlabel('Number of Components')
plt.ylabel('Eigenvalue')
plt.show()
```

### Components Output Table

Based on the variable load in the components output table below, the larger the number is within a
set, the more precision the set will yield for the collection within the principle component column.

**Example** Determining Variable Load for PC1

Based on the above table from PC1, the components listed will have the most precision as a

```python
#Components Output Table
loadings = pd.DataFrame(pca.components_.T, index=df[numeric_data].columns)
loadings
```
